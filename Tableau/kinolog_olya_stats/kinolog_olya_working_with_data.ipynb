```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('kinolog.csv', sep = ',')
```

```python
df1 = df.drop('Число часов трансляций', axis = 1)
```

```python
df1 = df.drop('Число часов трансляций', axis = 1)
```

```python
df2 = df1.drop(0)
```

```python
df2.head()
```

```python
df2 = df2.rename(columns = {'Контент':'content', 'Название видео':'title', 'Время публикации видео':'video_date', 'Средний процент просмотра (%)':'avg_viewing_percent', 'Средняя продолжительность просмотра':'avg_viewing_duration', 'Новые комментарии':'video_comment', 'Отметки "Не нравится"':'video_dislikes', 'Отметки "Нравится"':'video_likes', 'Просмотры':'video_views', 'Время просмотра (часы)':'viewing_time_hours', 'Подписчики':'subscribers', 'Показы':'shows', 'CTR для значков видео (%)':'ctr', })
```

```python
df2['video_date'] = pd.to_datetime(df2['video_date'], format="%b %d, %Y", errors='coerce')
```

```python
df2['video_date'] = pd.to_datetime(df2['video_date']).dt.strftime("%d.%m.%Y")
```

```python
df2['viewing_time_hours'] = df2['viewing_time_hours'].str.replace(',', '.')
```

```python
df2['avg_viewing_percent'] = df2['avg_viewing_percent'].str.replace(',', '.')
```

```python
df2['ctr'] = df2['ctr'].str.replace(',', '.')
```

```python
df2
```

```python
# Функция для применения логики
def determine_video_type(description):
    if '#' in description:
        return 'shorts'
    else:
        return 'full_video'
```

```python
# Примените функцию к колонке 'description' и создайте новую колонку 'video_type'
df2['video_type'] = df2['title'].apply(determine_video_type)
```

```python
df2
```

```python
df2.groupby(['video_type'] ,as_index = False)\
   .agg({'video_comment':'count'})
```
