```python
# 4. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)? Не стоит забывать, 
# что внутри месяца может быть не целое количество недель. Например, в ноябре 2021 года 4,28 недели. 
# И внутри метрики это нужно учесть. (8 баллов)
```

```python
# Cоберем датафрейм для анализа
df_order_plus_df_items = df_items.merge(df_order, on = 'order_id')
```

```python
df = df_order_plus_df_items.merge(df_id, on = 'customer_id')
df.shape
df.head()
```

```python
# Можем убрать пропущенные значения, заказы без времени передачи заказа в логистическую службу, 
# времени доставки заказа и времени подтверждения оплаты заказа не интересуют, 
# т.к. мы их не считаем за покупку.
df = df.dropna()
```

```python
# # для дальнейшей работы приведу все даты к формату Date, чтобы было проще работать с ними.
df.order_purchase_timestamp = pd.to_datetime(df.order_purchase_timestamp)
# df.order_approved_at = pd.to_datetime(df.order_approved_at)
# df.order_delivered_carrier_date = pd.to_datetime(df.order_delivered_carrier_date)
# df.order_delivered_customer_date = pd.to_datetime(df.order_delivered_customer_date)
# df.order_estimated_delivery_date = pd.to_datetime(df.order_estimated_delivery_date)
```

```python
# cчитаем количество недель в месяце

df['weeks'] = df.order_purchase_timestamp.dt.days_in_month / 7
```

```python
df
```

```python
df['order_purchase_timestamp_name'] = df.order_purchase_timestamp.dt.month_name()
# переводим дату в формат названия месяца
df
```

```python
# группируем данные для расчета среднего кол-ва выполненых заказов
customer_per_month  = df.query('order_delivered_customer_date != ""')\
    .groupby(['customer_unique_id','order_purchase_timestamp_name','weeks'], as_index=False)\
    .agg({'order_status':'count'})\
    .rename(columns={'order_purchase_timestamp_name':'month','order_status':'avg_quantity'}) 
customer_per_month.head()
```

```python
# выявляем среднее кол-во покупок
customer_per_month.avg_quantity = customer_per_month.avg_quantity / customer_per_month.weeks
```

```python
customer_per_month
```

```python
# Мы посчитали среднее количество заказов, но есть одно но, мы не учли, что уникальных
# customer_per_month меньше, чем shape таблицы, а это значит, что покупки одного пользователя
# посчитаны несколько раз, попробуем это исправить.
customer_per_month.customer_unique_id.nunique()
```

```python
# группируем данные для расчета среднего кол-ва выполненых заказов для уникальных пользователей
customer_per_month_new  = df.query('order_delivered_customer_date != ""')\
                            .groupby(['customer_unique_id'], as_index=False)\
                            .agg({'order_status':'count', 'weeks':'sum'})\
                            .rename(columns={'order_status':'quantity'})\
                            .sort_values('quantity')
customer_per_month_new['avg_quantity'] = customer_per_month_new.quantity/customer_per_month_new.weeks
customer_per_month_new
# отражает среднее в месяц в который была активность(покупка) к общему количеству месяцев.
```

```python
# Но и это ещё не все, на примере пользователя, видим, что order_item_id идёт по возрастающей
# это скорее всего позция в чеке, т.к время  у заказов с order_item_id  1,2,3 совпадает, отберем
# значения только с order_item_id = 1. Для того, чтобы посчитать реальное количество покупок.

# Строчку использовал для проверки, что все что идет ниже действительно верно отфильтровал, 
# и не посчитал покупку за одну дату за несколько покупок
df.query("customer_unique_id== 'ca77025e7201e3b30c44b472ff346268'") 
```

```python
real_quantity_per_user = df.query("order_item_id== '1'")\
                      .query('order_delivered_customer_date != ""')\
                      .groupby(['customer_unique_id'],as_index = False )\
                      .agg({'product_id' :'count'})\
                      .sort_values('product_id', ascending = False)\
                      .rename(columns={'product_id':'real_quantity'})
real_quantity_per_user
```

```python
# Cмержим с датафреймом, чтобы посчитать новое значение среднего
```

```python
real_avg_per_user = real_quantity_per_user.merge(customer_per_month_new, on = 'customer_unique_id')
real_avg_per_user['real_avg'] = real_avg_per_user.real_quantity/real_avg_per_user.weeks
real_avg_per_user
```

```python
# Также на мой взгляд, разумно посчитать среднее количество покупок следующим образом -
# (время жизни клиента/количество покупок), где время жизни клиента это текущая дата минус дата
# первой покупки.
# За текущую дату примем максимальную дату время создание заказа, предположив,что выгрузка
# была в эту дату.
```

```python
today = df.order_purchase_timestamp.max()
today 
```

```python
# Посчитаем сколько в неделях пользователь с нами (today -order_purchase_timestamp ) перед этим
# уберем время из дат, оставив только даты.
```

```python
df_with_us = df[['customer_unique_id','order_purchase_timestamp']]
df_with_us['order_purchase_timestamp'] = df_with_us['order_purchase_timestamp'].dt.date
today = today.date()
df_with_us['time_with_us'] = (today - df_with_us.order_purchase_timestamp)
df_with_us
```

```python
# Cреди customer_unique_id оставляем только те для которых время time_with_us - максимальное.
# так мы найдем время которое пользователь с нами исключая счетчик с новых заказов.
```

```python
max_time_with_us = df_with_us.groupby('customer_unique_id')['time_with_us']\
                              .max()\
                              .reset_index()\
                              .rename(columns = {'time_with_us':'real_time_with_us'})
max_time_with_us.head()
```

```python
max_time_with_us.dtypes
```

```python
# Переведем значения real_time_with_us в недели, и смержим с другим датафреймом чтобы посчитать 
# среднее количество покупок в неделю по месяцам в которые пользователь присутсвовал на сервисе.
# и посчитаем среднее
max_time_with_us['real_time_with_us'] = max_time_with_us['real_time_with_us'] / pd.Timedelta(1, 'W')
```

```python
real_2_avg = max_time_with_us.merge(real_avg_per_user, on = 'customer_unique_id')
real_2_avg['real_2_avg'] = real_2_avg.real_quantity/real_2_avg.real_time_with_us
real_2_avg.sort_values('real_2_avg')
```

```python
# Проверим что ничего не потеряли
df.query('order_delivered_customer_date != ""') \
    .customer_unique_id.nunique() == customer_per_month.customer_unique_id.nunique() == real_2_avg.customer_unique_id.nunique()
```

```python
# real_time_with_us появились нули, а в значениии real_2_avg inf, я думаю это вызвано, тем,
# что мы приняли дату выгрузки на сегодня - today, а в этот день были заказы, уберем их датафрейма,
# чтобы посчитать среднии значения и посмотреть график.
real_2_avg.describe()
```

```python
total_df = real_2_avg.query("real_2_avg!='inf'and real_2_avg!='NaN'")
total_df.describe()
# По той же причине максимальное значение очень большое, пользователь только зарегестрировался
# и уже сделал покупку 
```

```python
# Выводы: На мой взгляд смотреть среднее количество покупок по месяцам разумнее, как количество
#     покупок у пользователя разделенное на период его нахождения с нами(время с начала регистрации
#     по сегодняшний день) т.к. иначе мы получаем среднее количество покупок по неделям 
#     в месяц только в активные месяцы(когда была покупка), такая метрика тоже имеет место быть, 
#     но для сравнения активности пользователей на дистанции я бы пользовался все таки метрикой, 
#     которая показывает среднее количество покупок в неделю по месяцам в течение всего времени пока
#     пользователей с нами, например, если у какой - то группы пользователей снижается среднее 
#     количество покупок на дистанции, значит самое время им напомнить, что давно они ничего не
#     заказывали у нас. И real_2_avg среднее заметно ниже, чем у других метрих, т.к. мы смотрим
#     на большой дистанции.
    
```
